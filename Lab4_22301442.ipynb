```python
class Patient:      #DoublyNodeClass
  def __init__(self, id ,name, age, blood, next=None, prev=None):
    self.id = id
    self.name = name
    self.age = age
    self.blood = blood
    self.next = next
    self.prev = prev

class WRM:        #LinkedList
  def __init__(self):
    self.dh = Patient(None, None, None, None)
    self.dh.next = self.dh
    self.dh.prev = self.dh
    self.tail = self.dh

  def option(self, number):
    if number == 1:
      name = input("Name: ")
      id = int(input("ID: "))
      age = int(input("Age: "))
      blood = input("Blood grp: ")
      self.RegisterPatient(id,name,age,blood)
    elif number == 2:
      self.ServePatient()
    elif number == 3:
      self.CancelAll()
    elif number == 4:
      self.CanDoctorGoHome()
    elif number == 5:
      self.ShowAllPatient()
    elif number == 6:
      self.ReverseTheLine()
    else:
      self.exit()

  def RegisterPatient(self, id, name, age, blood):
    new_patient = Patient(id, name, age, blood, self.dh, self.tail)
    self.tail.next = new_patient
    self.dh.prev = new_patient
    self.tail = new_patient

    print("Successfully registered a patient.\n"
         "==========================\n")

    choice = int(input("Choose an option: "))
    self.option(choice)

  def ServePatient(self):
    if self.dh.next == self.dh:
      print("No patient left.")
    else:
      patient = self.dh.next
      print(patient.name,"has been served")
      next_patient = patient.next
      self.dh.next = next_patient
      next_patient.prev = self.dh

    patient = None

    choice = int(input("Choose an option: "))
    self.option(choice)

  def CancelAll(self):
    if self.dh.next == self.dh:
      print("No patient left. Already cancelled all.")
    else:
      self.dh.next = self.dh
      self.dh.prev = self.dh

    print("Successfully cancelled all patients.")
    choice = int(input("Choose an option: "))
    self.option(choice)

  def CanDoctorGoHome(self):
    if self.dh.next == self.dh and self.dh.prev == self.dh:
      print('Yes')
    else:
      print('No')

    choice = int(input("Choose an option: "))
    self.option(choice)

  def ShowAllPatient(self):
    if self.dh.next == self.dh:
      print("No patient left.")
    else:
      n = self.dh.next
      while n != self.dh:
         print(n.name, end = " ")
         n = n.next
    print()

    choice = int(input("Choose an option: "))
    self.option(choice)

  def ReverseTheLine(self):
    if self.dh.next == self.dh:
      print("Cannot reverse the line. There is no patient.")
      return
    else:                           #counting Nodes
      self.tail = self.dh.next
      count = 0
      while self.tail != self.dh:
        self.tail = self.tail.next
        count += 1

    cur_node = self.dh.next

    for i in range(0,count):        #so the code will run 3 times, if there are 3 nodes(patients).
      next_node = cur_node.next
      previous_node = cur_node.prev

      previous_node.prev = cur_node
      cur_node.next = previous_node

      if i == (count-1):
        pass
      else:
        cur_node = next_node
    self.dh.next = cur_node
    cur_node.prev = self.dh

    print("Successfully reversed the line.")
    choice = int(input("Choose an option: "))
    self.option(choice)

  def exit(self):
    print("bye")
```
